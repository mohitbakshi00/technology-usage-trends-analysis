# Web Scraping the programming languages data


## Extracting the information from the web site



```python
#this url contains the data about the popular programmingLanguages and average annual salary
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

```python
# importing neccessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
response = requests.get(url).text
response
```

```python
soup = BeautifulSoup(response,'html5lib') # Creating a soup object
soup
```

Scraping the `Language name` and `annual average salary`.


```python
table = soup.find('table')
for row in table.find_all('tr'):
    cols = row.find_all('td') 
    print(cols[1].get_text()," ---> ",cols[3].get_text())            
```

Saving the scrapped data into a file named *popular-languages.csv* for futher analysis

```python
a=[]
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td') 
    a.append([cols[1].get_text(),cols[3].get_text()])

df= pd.DataFrame(a,columns=[['Language', 'Average Annual Salary']])


df.reset_index(drop=True,inplace=True)
df.to_csv('popular-languages.csv')
```
