```python
#Connecting to the database
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.set_option('display.max_columns',None)
```

## listing all the tables


```python
QUERY = """SELECT name as Table_Name FROM 
sqlite_master WHERE  type = 'table'"""   

pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
df=pd.read_sql_query(QUERY,conn)
df.head()
```

```python
table_name = 'master'  # describing the table

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
QUERY = """
SELECT* FROM master """
df=pd.read_sql_query(QUERY,conn)
df.head()
```

```python
#plotting the box plot of age
fig = px.box(df,y="Age") 
fig.layout.template='plotly_dark'
fig.show()
```

## Visualizing relationships in data


```python
df['Age'].dropna(inplace=True)
```

```python
# scatter plot between age and working hours
fig = px.scatter(df,x='Age',y="WorkWeekHrs",color='WorkWeekHrs')
fig.layout.template='plotly_dark'
fig.show()
```

```python
df1 = df[df['Age'].notna()]
```

```python
# bubble plot of WorkWeekHrs and CodeRevHrs, with Age
fig = px.scatter(df1,x='WorkWeekHrs',y="CodeRevHrs",color='Age',size='Age')
fig.layout.template='plotly_dark'
fig.show()
```

## Visualizing composition of data


```python
query="""select DatabaseDesireNextYear,count(DatabaseDesireNextYear) as cnt  
from DatabaseDesireNextYear group by DatabaseDesireNextYear order by cnt desc limit 5 """

df1=pd.read_sql_query(query,conn)
fig = px.pie(df1,values='cnt', names='DatabaseDesireNextYear', title='top 5 databases')
fig.layout.template='plotly_dark'
fig.show()
```

```python
#stacked chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
df3=df.query('Age>=30 and Age <=35')[['Age','WorkWeekHrs','CodeRevHrs']]
fig=px.bar(df3,x='Age',y=['WorkWeekHrs','CodeRevHrs'],barmode='stack')
fig.layout.template='plotly_dark'
fig.update_traces(marker_line_width=0)
fig.show()
```

## Visualizing comparison of data


```python
#Ploting the median ConvertedComp for all ages from 45 to 60
df2=df.query('Age>=45 and Age <=60')[['Age','ConvertedComp']]
df2=df2.groupby('Age').aggregate('median')
df2.reset_index(inplace=True)
```

```python
fig = px.line(df2,x='Age',y='ConvertedComp')
fig.layout.template="plotly_dark"
fig.show()
```

```python
df2=df.query('Age>=25 and Age <=30')[['Age','ConvertedComp']]
df2=df2.groupby('Age').aggregate('median')
df2.reset_index(inplace=True)
fig = px.line(df2,x='Age',y='ConvertedComp')
fig.layout.template="plotly_dark"
fig.show()
```

```python
query = """ select MainBranch ,count(*) as cnt from master group by MainBranch"""
df=pd.read_sql_query(query,conn)
df
```

```python
#Creating the horizontal bar chart using of MainBranch
df['MainBranch']=np.where(df['MainBranch']=='I am not primarily a developer, but I write code sometimes as part of my work','not a developer',np.where(df['MainBranch']=='I am a developer by profession','developer',None))
fig = px.bar(df,y='MainBranch',x='cnt',color='MainBranch')
fig.layout.template='plotly_dark'
fig.show()
```
