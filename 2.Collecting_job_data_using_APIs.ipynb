# **Collecting Job Data Using APIs**


```python
import requests 
```

Collecting the number of job postings for the following languages using the API:

*   C
*   C#
*   C++
*   Java
*   JavaScript
*   Python
*   Scala
*   Oracle
*   SQL Server
*   MySQL Server
*   PostgreSQL
*   MongoDB


```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/datasets/githubposting.json"
response = requests.get(url)
data = response.json()
print(data)
print(data[0]['A'])
```

```python
# function to get the number of jobs for the given technology
def get_number_of_jobs(technology):
    number_of_jobs = 0
    for i in data:
        if(i['A']==technology):
                number_of_jobs = int(i['B'])
                break
    return technology,number_of_jobs
```

```python
print(get_number_of_jobs('Python'))
print(get_number_of_jobs('Java'))
print(get_number_of_jobs('C'))
```

### Storing the results in an excel file


```python
from openpyxl import Workbook        
wb=Workbook()                        
ws=wb.active                         
ws.append(['Technology','number_of_jobs'])   
for i in data[1:]:
    ws.append([i['A'],int(i['B'])])
wb.save("jobs.xlsx")  
```
